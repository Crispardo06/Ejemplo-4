```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# Parámetros del sistema
params = {
    "numerador": [3],         # Numerador de la función de transferencia
    "denominador": [3, 6, 9], # Denominador (s^2 + 2s + 1)
    "num_puntos": 50         # Número de puntos en la simulación
}

# Crear la función de transferencia
system = ct.TransferFunction(params["numerador"], params["denominador"])

# Obtener la respuesta al escalón
t, y = ct.step_response(system, T=np.linspace(0, 10, params["num_puntos"]))

# Mostrar datos de entrada
print("Datos de Entrada:")
print(f"Numerador: {params['numerador']}")
print(f"Denominador: {params['denominador']}")
print(f"Intervalo de tiempo: [0, 10]\n")

# Mostrar datos de salida (primeros 10 valores)
print("Datos de Salida: (Primeros 10 puntos)")
for i in range(10):
    print(f"Tiempo: {t[i]:.2f} s, Salida: {y[i]:.4f}")

# Personalización del color de la gráfica
color_grafica = "green"  # Cambia el color aquí ('red', 'blue', etc.)

# Graficar la respuesta al escalón
plt.plot(t, y, color=color_grafica, label="Salida del sistema")
plt.title("Respuesta al Escalón de un Sistema de Segundo Orden")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida")
plt.legend()
plt.grid()
plt.show()
```
